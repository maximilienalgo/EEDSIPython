```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

```

```python
sales_video_games = pd.read_csv("./vgsales-12-4-2019.csv")
sales_video_games.info()
```

```python
sales_video_games.head()
```

```python
sns.catplot(data=sales_video_games, x="Global_Sales", y="Platform", height=10)
```

```python
sales_video_games.describe()
```

```python
plt.figure(figsize=(30,10))
plt.bar(sales_video_games['Platform'].value_counts().index,sales_video_games['Platform'].value_counts())
```

```python
sales = sales_video_games[['Genre', 'NA_Sales', 'PAL_Sales', 'JP_Sales', 'Other_Sales']]
sales = sales.groupby('Genre').sum()
sales.plot.bar(figsize=(20,10))
plt.show()
```

```python
sales_video_games5 = sales_video_games.sort_values(by=['Total_Shipped'],ascending=False)
top5 = sales_video_games5[['Total_Shipped', 'Name']]
top5 =top5.head()
sns.catplot(x="Name", y='Total_Shipped',kind="bar", data=top5, height=10)
```

```python
sns.relplot(x="Year", y="Global_Sales", data = sales_video_games, kind="line")
```
